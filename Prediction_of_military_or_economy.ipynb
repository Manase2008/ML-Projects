```python
import pandas as pd
```

```python
hey = pd.read_csv('armenpress_army.csv')
hey1 = pd.read_csv('armenpress_economy.csv')
# upload from your computer the files
hey['type'] = 'military'
hey1['type'] = 'economy'
```

```python
merged_data = pd.merge(hey, hey1, how = 'outer')
merged_data
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

merged_data["type"] = encoder.fit_transform(merged_data["type"])
```

```python
random_merged_data = merged_data.sample(frac=1, random_state=42).reset_index(drop=True)
random_merged_data
```

```python
X = random_merged_data.drop('type', axis=1)
y = random_merged_data['type']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words='english', ngram_range=(1,1) )
X_train_counts = vectorizer.fit_transform(X_train['article_paragraph'])
X_test_counts = vectorizer.transform(X_test['article_paragraph'])
```

```python
from sklearn.naive_bayes import MultinomialNB

# Initialize the model
nb_classifier = MultinomialNB()

# Train the model
nb_classifier.fit(X_train_counts, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Make predictions
predictions = nb_classifier.predict(X_test_counts)

# Check accuracy
print("Accuracy:", accuracy_score(y_test, predictions))
```

```python
new_data = ['The government of Armenia is seriously keeping track of its water resources', 'There has been signed a serious military contract between USA and Armenia']
new_data_vector = vectorizer.transform(new_data)
predictions = nb_classifier.predict(new_data_vector)
predictions
```
