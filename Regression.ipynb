```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('california_housing_train.csv')
data
```

```python
test = pd.read_csv('california_housing_test.csv')
test
```

```python
X = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_test = test.drop(['median_house_value'], axis=1)
y_test = test['median_house_value']
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_scaled, y)
```

```python
prediction = model.predict(X_test_scaled)
```

```python
# Metrics
r2 = r2_score(y_test, prediction)
mae = mean_absolute_error(y_test, prediction)
mse = mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)

print('r2 accuracy:', r2)
print('mae accuracy:', mae)
print('mse accuracy:', mse)
print('rmse accuracy:', rmse)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data = pd.read_excel('synthetic_loan_risk.xlsx')
data
```

```python
categorical_cols = ['EducationLevel', 'House', 'EmploymentType', 'MaritalStatus', 'Region', 'YearsJob']

# Apply LabelEncoder to these columns
for col in categorical_cols:
    if col in data.columns:
        data[col] = encoder.fit_transform(data[col])
```

```python
X = data.drop('Default', axis=1)
y = data['Default']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=100000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
data = pd.read_csv('california_housing_train.csv')
test = pd.read_csv('california_housing_test.csv')
```

```python
X = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_test = test.drop(['median_house_value'], axis=1)
y_test = test['median_house_value']
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

model = SVR(
    kernel="rbf",
    C=1.0,
    epsilon=0.1
)

model.fit(X_scaled, y)

predictions = model.predict(X_test_scaled)
r2 = r2_score(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print("R-squared:", r2)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```
