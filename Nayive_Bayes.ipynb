```python
import pandas as pd
```

```python
data = pd.read_excel('synthetic_loan_risk_1660x15features.xlsx')
data.tail()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
categorical_cols = ['EducationLevel', 'EmploymentType', 'MaritalStatus', 'Region', 'YearsJob', 'House']

# Apply LabelEncoder to these columns
for col in categorical_cols:
    if col in data.columns:
        data[col] = encoder.fit_transform(data[col])
```

```python
# Now, redefine X and y from the fully preprocessed data
X = data.drop("Default", axis=1)
y = data["Default"]

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Perform the train-test split (80% train, 20% test)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize the Gaussian Naive Bayes model
naive_bayes_model = GaussianNB()

# Train the model on the training data
naive_bayes_model.fit(X_train, y_train)

print("Gaussian Naive Bayes model trained successfully!")
```

```python
# Make predictions on the test set using the Naive Bayes model
nb_prediction = naive_bayes_model.predict(X_test)

print("Predictions made using Naive Bayes:")
print(nb_prediction)
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the Naive Bayes model
nb_accuracy = accuracy_score(y_test, nb_prediction)

print(f"Naive Bayes Model Accuracy: {nb_accuracy:.4f}")
```
