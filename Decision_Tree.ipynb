```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.tree import export_graphviz, DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data = pd.read_csv('train.csv')
data = data.dropna(how='any')
categorical_cols = ['PassengerId', 'Ticket', 'Pclass', 'Name', 'Sex', 'Fare', 'Cabin', 'Embarked']

for col in categorical_cols:
    if col in data.columns:
        data[col] = encoder.fit_transform(data[col])
```

```python
y = data.Survived
X = data.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Ticket', 'Fare'], axis=1)
X
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model1 = DecisionTreeClassifier(criterion='entropy', max_depth=2)
model1.fit(X_train, y_train)
prediction1 = model1.predict(X_test)
accuracy_score(prediction1, y_test)
```
