```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('synthetic_loan_risk_1660x15features.xlsx')
data.tail()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data["House"] = encoder.fit_transform(data["House"])
```

```python
# Identify remaining categorical columns (excluding 'Default' which is the target)
categorical_cols = ['EducationLevel', 'EmploymentType', 'MaritalStatus', 'Region', 'YearsJob']

# Apply LabelEncoder to these columns
for col in categorical_cols:
    if col in data.columns:
        data[col] = encoder.fit_transform(data[col])

display(data.head())
```

```python
# Now, redefine X and y from the fully preprocessed data
X = data.drop("Default", axis=1)
y = data["Default"]

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Perform the train-test split (80% train, 20% test)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=40)

model.fit(X_train, y_train)
```

```python


prediction = model.predict(X_test)

print(prediction)
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, prediction)

print(f"KNN Model Accuracy: {accuracy:.4f}")
```
